```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("projectData/ECG5000_TEST.txt", sep='  ', engine='python', header=None)
```

```python
data
```

```python
data = data.add_prefix('c')
```

```python
data.head()
```

```python
count = data['c0'].value_counts()
count
```

```python
normal_count = count[1.0]
anomaly_count = count[count.index.isin([2.0, 3.0, 4.0, 5.0])].sum()
print(f"Amount of normal cases: {normal_count}")
print(f"Amount of anomalous cases: {anomaly_count}")
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
print(f"Shape of the data: {data.shape}")
```

```python
y = data['c0']
x = data.drop('c0', axis=1)
X_train = x[:3600]
y_train = y[:3600]
X_test = x[3600:]
y_test = y[3600:]
```

```python
print(f"Shapes: X_train: {X_train.shape}, y_train: {y_train.shape}, X_test: {X_test.shape}, y_test: {y_test.shape}")
```

```python
scaler = MinMaxScaler()
scaled_data = scaler.fit(X_train)
trained_data_scaled = scaler.transform(X_train)
test_data_scaled = scaler.transform(X_test)
```

```python
X_train_anomalies = trained_data_scaled[y_train != 1.0]
X_train_normal = trained_data_scaled[y_train == 1.0]
X_test_anomalies = test_data_scaled[y_test != 1.0]
X_test_normal = test_data_scaled[y_test == 1.0]
```

```python
plt.plot(X_train_normal[0], label='normal 1')
plt.plot(X_train_normal[1], label='normal 2')
plt.legend()
plt.show()
```

```python
plt.plot(X_train_anomalies[0], label='anomaly 1')
plt.plot(X_train_anomalies[1], label='anomaly 2')
plt.legend()
plt.show()
```
